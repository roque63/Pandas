# Pandas

Su objetivo es ser el bloque de alto nivel para **realizar análisis de datos del mundo real** en Python. Además, tiene el objetivo más amplio de convertirse en la herramienta de manipulación/análisis de datos de fuente abierta más poderosa y flexible disponible en cualquier idioma .

Las dos estructuras de datos principales de pandas, Series(unidimensional) y DataFrame(bidimensional), manejan la gran mayoría de los casos de uso típicos en finanzas, estadísticas, ciencias sociales y muchas áreas de la ingeniería. Para los usuarios de R, DataFrame proporciona todo lo que ofrece R data.frame y mucho más. pandas se basa en NumPy y está diseñado para integrarse bien en un entorno informático científico con muchas otras bibliotecas de terceros.

Operaciones disponibles en Pandas:

* Fácil manejo de datos faltantes (representados como NaN) en punto flotante y datos de punto no flotante

* Mutabilidad de tamaño: las columnas se pueden insertar y eliminar del DataFrame.

* Alineación de datos automática y explícita : los objetos se pueden alinear explícitamente con un conjunto de etiquetas, o el usuario puede simplemente ignorar las etiquetas y dejar que Series, DataFrame, etc. alineen automáticamente los datos para usted en los cálculos.

* Agrupar por funcionalidad potente y flexible para realizar operaciones de división, aplicación y combinación en conjuntos de datos, tanto para agregar como para transformar datos

* Facilite la conversión de datos desiguales e indexados de manera diferente en otras estructuras de datos de Python y NumPy en objetos DataFrame

* Segmentación inteligente basada en etiquetas , indexación sofisticada y creación de subconjuntos de grandes conjuntos de datos

* Combinación y unión de conjuntos de datos intuitivos

* Transformación y rotación flexibles de conjuntos de datos

* Etiquetado jerárquico de ejes (es posible tener varias etiquetas por marca)

* Robustas herramientas de E/S para cargar datos desde archivos planos (CSV y delimitados), archivos de Excel, bases de datos y guardar/cargar datos desde el formato ultrarrápido HDF5

* Funcionalidad específica de series de tiempo : generación de rango de fechas y conversión de frecuencia, estadísticas de ventana móvil, cambio de fecha y retraso.  

Estructuras de datos

1. **Series**:    Matriz de tipado homogéneo etiquetada en 1D

2. **DataFrame**:    Estructura tabular de tamaño mutable etiquetada en 2D general con columna tipificada potencialmente heterogénea

¿Cómo filtro renglones específicos de un DataFrame? ¶
![texto del vínculo](https://pandas.pydata.org/pandas-docs/stable/_images/03_subset_rows.svg)

🧐    Me interesan los pasajeros mayores de 35 años.
Para seleccionar renglones, utilice una condición dentro de los corchetes de selección [].

La condición dentro de los corchetes de selección verifica qué renglones de la columna "Age" tiene un valor mayor que 35:
* **titanic["Age"] > 35**

La salida de la expresión condicional ( >, pero también ==, !=, <, <=,… funcionaría) es en realidad un panda Series de valores booleanos (ya sea True o False) con el mismo número de renglones que el DataFrame original. Estos valores booleanos Series se pueden usar para filtrar renglones del DataFrame colocándolos entre los corchetes de selección [].
Así solo se seleccionarán las filas para las que el valor es True .

Sabemos desde antes que el DataFrame Titanic original consta de 891 renglones.
Revisemos la cantidad de renglones que satisfacen la condición al verificar el atributo __shape__ del __above_35__: 217 pasajeros tenian una edad : a 35.



```python
above_35 = titanic[titanic["Age"] > 35]
print(above_35.head())
print(above_35.shape)
```

🤓 Me interesan los pasajeros del Titanic de cabina clase 2 y 3.🚢


Para filtrar los renglones se utiliza la función condicional **isin( )** dentro de los corchetes de selección []. En este caso, la condición dentro de los corchetes de selección verifica qué renglones de la columna tiene valores 2 o 3.

* class_23 = titanic["Pclass"].isin([2, 3])

Lo anterior es equivalente a filtrar por renglones para las cuales la "Pclass" es 2 o 3 y combinar las dos declaraciones con un | operador or:

* class_23 = titanic[(titanic["Pclass"] == 2) | (titanic["Pclass"] == 3)]

Nota: Al combinar varias condicionales, cada condición debe estar entre paréntesis ().
Además, no se pueden usar el or/and
se debe usar el operador **|** para el or y operador **&** para el and .


```python
class_23 = titanic[titanic["Pclass"].isin([2, 3])]
print(class_23)
class_23 = titanic[(titanic["Pclass"] == 2) | (titanic["Pclass"] == 3)]
class_23
```

¿Cómo selecciono renglones y columnas de un DataFrame?
![texto del vínculo](https://pandas.pydata.org/pandas-docs/stable/_images/03_subset_columns_rows.svg)

En este caso, se crea un subconjunto de renglones y columnas de una sola vez y ya []no es suficiente usar corchetes de selección.
Los operadores loc/ iloc son obligatorios antes de los corchetes de selección [].
* Al usar loc/iloc, la parte anterior a la coma son los renglones y la parte posterior a la coma son las columnas que desea seleccionar.

* Cuando utilices nombres en las columnas, las etiquetas de las filas o una expresión de condición, utiliza el loc antes de los corchetes de selección
nombre_DataFrame.loc[,].
* Tanto para la parte anterior como posterior a la coma, puede usar una sola etiqueta, una lista de etiquetas, una porción de etiquetas, una expresión condicional o dos puntos.
* El uso de dos puntos especifica que desea seleccionar todas las filas o columnas.

* 📌   ¿Cómo obtener los nombres de los pasajeros que edad mayor a 35 años?




```python
adult_names = titanic.loc[titanic["Age"] > 35, "Name"]
adult_names.head()
```

para calcular la  concentración de la estación de Londres en mg/m

Para crear una nueva columna, usa los [ ] con el nombre de la nueva columna en el lado izquierdo de la asignación.

* Nota : El cálculo de los valores se realiza element_wise . Esto significa que todos los valores en la columna dada se multiplican por el valor 1.882 a la vez. ¡No necesita usar un ciclo para iterar cada una de los renglones.

# Agregando estadísticas agrupadas por categoría ¶
![texto del vínculo](https://pandas.pydata.org/docs/_images/06_groupby.svg)


📌 ¿Cuál es la edad promedio de los pasajeros masculinos y femeninos del Titanic?

Como nuestro interés es la edad promedio para cada género, primero se hace una subselección en estas dos columnas:  posteriormente, se aplica el método en la columna para hacer un grupo por categoría. La edad promedio para cada sexo se calcula y se devuelve.

* **titanic[["Sex", "Age"]].groupby("Sex").mean()**



```python
genero_edad_df = titanic[["Sex", "Age"]].groupby("Sex")
```

```python
titanic[["Sex", "Age"]].groupby("Sex").mean()
```

El método groupby se utiliza para soportar este tipo de operaciones. Más general, esto encaja en el split-apply-combine patrón más general:

* Dividir los datos en grupos

* Aplicar una función a cada grupo de forma independiente

* Combinar los resultados en una estructura de datos

Los pasos de aplicar y combinar normalmente se realizan juntos en pandas.

En el ejemplo anterior, seleccionamos explícitamente las 2 columnas primero.
Si no, mean  se aplica a cada columna que contiene columnas numéricas:

No tiene mucho sentido obtener el valor promedio de Pclass. si solo nos interesa la edad promedio para cada género, la selección de columnas (corchetes []como de costumbre) también se admite en los datos agrupados:

* titanic.groupby("Sex")["Age"].mean()


![texto del vínculo](https://pandas.pydata.org/docs/_images/06_groupby_select_detail.svg)

```python
titanic.groupby("Sex").mean()
```

```python
titanic.groupby("Sex")["Age"].mean()
```
